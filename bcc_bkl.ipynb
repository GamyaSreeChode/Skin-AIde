```python
import os, glob, time, warnings, numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model, mixed_precision
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix

mixed_precision.set_global_policy('mixed_float16')
tf.random.set_seed(42); np.random.seed(42)
AUTOTUNE = tf.data.AUTOTUNE

# ── Config ───────────────────────────────────────────────────
model_dir   = "/kaggle/input/models/trizenventuresllp/disease/keras/default/1"
keras_files = glob.glob(os.path.join(model_dir, "*.keras"))

CONFIG = {
    "image_size": 300, "batch_size": 32, "epochs": 30, "lr": 5e-4,
    "data_dir":      "/kaggle/input/datasets/kmader/skin-cancer-mnist-ham10000",
    "metadata_path": "/kaggle/input/datasets/kmader/skin-cancer-mnist-ham10000/HAM10000_metadata.csv",
    "base_model_path": keras_files[0],
    "specialist_save": "/kaggle/working/specialist_bcc_bkl",
}

CLASS_NAMES  = ['akiec','bcc','bkl','df','mel','nv','vasc']
WEAK_CLASSES = ['bcc','bkl']
WEAK_IDX     = [CLASS_NAMES.index(c) for c in WEAK_CLASSES]
IMG_SIZE     = CONFIG["image_size"]

# ── Load old model + feature extractor ───────────────────────
old_model = keras.models.load_model(CONFIG["base_model_path"])
old_model.trainable = False

sub_model  = old_model.layers[-3]
feat_layer = None
for l in reversed(sub_model.layers):
    try:
        shape = l.output.shape
        if len(shape) == 2 and shape[-1] not in [7, None]:
            feat_layer = l; break
    except: continue

feature_extractor = Model(inputs=sub_model.input, outputs=feat_layer.output)
feature_extractor.trainable = False
print(f"Feature extractor: {feature_extractor.output_shape}")

# ── Data ─────────────────────────────────────────────────────
df = pd.read_csv(CONFIG["metadata_path"])
img_paths = {}
for root, _, files in os.walk(CONFIG["data_dir"]):
    for f in files:
        if f.endswith(('.jpg','.jpeg','.png')):
            img_paths[os.path.splitext(f)[0]] = os.path.join(root, f)

df['image_path'] = df['image_id'].map(img_paths)
df = df.dropna(subset=['image_path']).reset_index(drop=True)
df['label'] = df['dx'].map({c:i for i,c in enumerate(CLASS_NAMES)})

df_train, df_temp = train_test_split(df, test_size=0.30, stratify=df['dx'], random_state=42)
df_val,   df_test = train_test_split(df_temp, test_size=0.50, stratify=df_temp['dx'], random_state=42)

# bcc & bkl only
s_train = df_train[df_train['dx'].isin(WEAK_CLASSES)].copy()
s_val   = df_val[df_val['dx'].isin(WEAK_CLASSES)].copy()
s_train['spec_label'] = s_train['dx'].map({'bcc':0,'bkl':1})
s_val['spec_label']   = s_val['dx'].map({'bcc':0,'bkl':1})

s_train = pd.concat([
    g.sample(3000, replace=len(g)<3000, random_state=42)
    for _, g in s_train.groupby('dx')
]).sample(frac=1, random_state=42).reset_index(drop=True)

print(f"Train:{len(s_train)} | Val:{len(s_val)} | Test:{len(df_test)}")

# ── tf.data ──────────────────────────────────────────────────
@tf.function
def load(path, label, augment=False):
    img = tf.io.read_file(path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.float32) / 255.0
    if augment:
        img = tf.image.random_flip_left_right(img)
        img = tf.image.random_flip_up_down(img)
        img = tf.image.random_brightness(img, 0.3)
        img = tf.image.random_contrast(img, 0.7, 1.3)
        img = tf.image.random_saturation(img, 0.7, 1.3)
        img = tf.image.random_crop(img, [int(IMG_SIZE*0.85), int(IMG_SIZE*0.85), 3])
        img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    return img, label

def make_ds(paths, labels, augment=False, shuffle=False):
    ds = tf.data.Dataset.from_tensor_slices((paths, labels.astype(np.int32)))
    if shuffle: ds = ds.shuffle(len(paths), seed=42)
    ds = ds.map(lambda p,l: load(p,l,augment), num_parallel_calls=AUTOTUNE)
    return ds.batch(CONFIG["batch_size"]).prefetch(AUTOTUNE)

train_ds     = make_ds(s_train['image_path'].values, s_train['spec_label'].values, augment=True, shuffle=True)
val_ds       = make_ds(s_val['image_path'].values,   s_val['spec_label'].values)
full_test_ds = make_ds(df_test['image_path'].values, df_test['label'].values)

# ── Specialist model ──────────────────────────────────────────
inputs   = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
features = feature_extractor(inputs, training=False)
x   = layers.Dense(256, activation='relu')(features)
x   = layers.BatchNormalization()(x)
x   = layers.Dropout(0.5)(x)
x   = layers.Dense(128, activation='relu')(x)
x   = layers.Dropout(0.4)(x)
out = layers.Dense(2, activation='softmax', dtype='float32')(x)

specialist = Model(inputs, out)
specialist.compile(optimizer=keras.optimizers.Adam(CONFIG["lr"]),
                   loss='sparse_categorical_crossentropy', metrics=['accuracy'])

cw = dict(enumerate(compute_class_weight('balanced',
     classes=np.array([0,1]), y=s_train['spec_label'].values)))

callbacks = [
    ModelCheckpoint(CONFIG["specialist_save"]+".weights.h5", monitor='val_accuracy',
                    save_best_only=True, save_weights_only=True, mode='max', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-7, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True, verbose=1),
]

# ── Train ─────────────────────────────────────────────────────
start = time.time()
h = specialist.fit(train_ds, validation_data=val_ds, epochs=CONFIG["epochs"],
                   class_weight=cw, callbacks=callbacks, verbose=1)
print(f"Done in {(time.time()-start)/60:.1f} min")

# ✅ Save full model
specialist.load_weights(CONFIG["specialist_save"]+".weights.h5")
specialist.save(CONFIG["specialist_save"]+".keras")
print(f"✅ Saved: {CONFIG['specialist_save']}.keras")

# ── Hybrid Predict ────────────────────────────────────────────
# bcc & bkl → specialist | everything else → old model
old_preds  = old_model.predict(full_test_ds, verbose=0)
spec_preds = specialist.predict(full_test_ds, verbose=0)

y_true = df_test['label'].values
y_pred = np.array([
    WEAK_IDX[np.argmax(spec_preds[i])] if np.argmax(old_preds[i]) in WEAK_IDX
    else np.argmax(old_preds[i])
    for i in range(len(old_preds))
])

# ── Results ───────────────────────────────────────────────────
print(classification_report(y_true, y_pred, target_names=CLASS_NAMES))
print(f"Overall Accuracy: {np.mean(y_pred==y_true)*100:.2f}%")
for i, name in enumerate(CLASS_NAMES):
    mask = y_true == i
    acc  = np.mean(y_pred[mask]==y_true[mask]) if mask.sum()>0 else 0
    src  = "SPECIALIST" if name in WEAK_CLASSES else "OLD MODEL"
    print(f"  [{src}] {'OK' if acc>=0.90 else 'LOW'} {name}: {acc*100:.2f}%")

# ── Plot ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14,5))
axes[0].plot(h.history['accuracy'],     label='Train')
axes[0].plot(h.history['val_accuracy'], label='Val')
axes[0].axhline(0.90, color='r', linestyle='--', label='90% target')
axes[0].set_title('Specialist (bcc & bkl)'); axes[0].legend()
cm = confusion_matrix(y_true, y_pred).astype(float)
sns.heatmap(cm/cm.sum(axis=1,keepdims=True), annot=True, fmt='.2f',
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES, ax=axes[1])
axes[1].set_title('Hybrid Confusion Matrix')
plt.tight_layout()
plt.savefig('/kaggle/working/results.png', dpi=120)
plt.show()
```

```python
from tensorflow import keras
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import glob, os

# ── Load models ───────────────────────────────────────────────
model_dir   = "/kaggle/input/models/trizenventuresllp/disease/keras/default/1"
keras_files = glob.glob(os.path.join(model_dir, "*.keras"))

old_model  = keras.models.load_model(keras_files[0])
specialist = keras.models.load_model("/kaggle/working/specialist_bcc_bkl.keras")

CLASS_NAMES  = ['akiec','bcc','bkl','df','mel','nv','vasc']
WEAK_IDX     = [1, 2]  # bcc, bkl

# ── Test image ────────────────────────────────────────────────
img_path = "/kaggle/input/datasets/kmader/skin-cancer-mnist-ham10000/HAM10000_images_part_1/ISIC_0027990.jpg"

img = image.load_img(img_path, target_size=(300, 300))
arr = np.expand_dims(image.img_to_array(img) / 255.0, axis=0)

# Predictions
old_probs  = old_model.predict(arr,  verbose=0)[0]
spec_probs = specialist.predict(arr, verbose=0)[0]
old_class  = np.argmax(old_probs)

# Routing
final = WEAK_IDX[np.argmax(spec_probs)] if old_class in WEAK_IDX else old_class

# ── Results ───────────────────────────────────────────────────
print(f"Old model said : {CLASS_NAMES[old_class]} ({old_probs[old_class]*100:.2f}%)")
if old_class in WEAK_IDX:
    print(f"Specialist said: {CLASS_NAMES[final]} ({spec_probs[np.argmax(spec_probs)]*100:.2f}%)")
    print(f"Final prediction: {CLASS_NAMES[final]} ← specialist used")
else:
    print(f"Final prediction: {CLASS_NAMES[final]} ← old model used")

# ── Show image ────────────────────────────────────────────────
plt.figure(figsize=(5,5))
plt.imshow(image.load_img(img_path))
plt.title(f"Predicted: {CLASS_NAMES[final]}", fontsize=14)
plt.axis('off')
plt.show()
```
